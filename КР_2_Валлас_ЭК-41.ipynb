# K-Means Clustering

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Using the silhouette method to find the optimal number of clusters
from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 6):
  kmeans = KMeans(n_clusters = k).fit(df)  
  preds = kmeans.fit_predict(df)
  sil.append(silhouette_score(df, preds, metric = 'euclidean'))


plt.plot(range(2, 6), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
 print(str(i+2) +":"+ str(sil[i]))
```

Рекомендовано выделить 2 кластера, но мы выделим 4

```python
# Fitting K-Means to the dataset
km = KMeans(n_clusters = 4, init = 'k-means++', random_state = 0).fit_predict(df)
kms = KMeans(n_clusters = 4, random_state = 0).fit(df)
pd.DataFrame(kms.cluster_centers_, columns = df.columns).round(3)
```

В первом кластере: неуспешные торг. точки, небольшой общей площади, с небольшим кол-вом сотрудников, социального типа, небольшим кол-вом чеков; без наличия Магазина моб. связи, Продуктового и Вещевого рынка, но с остановкой общ. транспорта; остальные показатели не выделяются.
Во втором: успешные торг. точки, с большим кол-вом сотрудников, чеков и ср. кол-вом единиц в чеке, премиального типа, с большим населением; рынков, магазина моб. связи и  остановок Общ. Тр. нет.
Третий кластер похож на второй, но в нем есть продуктовые и вещевые рынки, а также магазины моб. связи и остановка общ. транспорта. 
А четвертый кластер похож на первый, но там больше общая площадь, кол-во сотрудников и чеков. Торговые точки стандартного типа.
